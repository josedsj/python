```python
def wow(palabra):
    perms=[]
    wow_finales=[]
    for i in range(3, len(palabra)+1):
        perms+=(["".join(x) for x in permutations(palabra,i)])
    for i in perms:
        for j in dict:
            if(i==j):
                wow_finales+=[i]
    wow_finales=list(set(wow_finales))
    print(sorted(wow_finales, key=len))
```

```python
wow("hola")
```

```python
wow("trece")
```
